```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# plotting single line

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show() 
```

# multiple lines

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

# subplots

```python

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.subplot(2, 1, 1) 

#first plot
plt.plot(x, y_sin)
plt.title('Sine')

# second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```

Specifying axis

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
plt.axis([3,7,0,2])
plt.plot(x, y)
plt.show() 
```
