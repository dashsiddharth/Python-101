# Variables

Variables are nothing but reserved memory locations to store values. As I mentioned earlier, Python features a <b> dynamic type system</b> that means you don't have specify data type each time you create a new variable.

Python has five standard data types : 
<ul>
    <li> <b> Number </b></li>
    <li> <b> String </b></li>
    <li> <b> List </b></li>
    <li> <b> Tuple </b></li>
    <li> <b> Dictionary </b></li>
</ul>

```python
num1 = 1234                                                   #int
num2 = 123.45                                                 #float
char1 = 'S'                                                    #char
string1 = "College of Agricultural Engineering and Technology"  #string
str1 = None                                                   #denotes the lack of value
bool1 = True                                                 #boolean
```

```python
print(string1)
print(num1)
print(num2)
print(char1)
```

```python
print('col '*5)
print('col '+ '5')
```

```python
#multiple assignment

a,b,c = 's', 123, "Python"
print(a)
print(b)
print(c)
```

```python
#other print formats
str1 = "Sidd"
balance = 12000
print(" Hello %s your salary is %d" % (str1, balance) )
```

Remember
<ul>
    <li> %s -> string </li>
    <li> %d -> Integer</li>
     <li> %f -> Float </li>
<li> %o -> Octal </li>
<li> %x -> Hexadecimal </li>
<li> %e -> exponential </li>
</ul>

```python
print("Hello {} your salary is {}".format(str1, balance))
```
